```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Housing.csv')
df.head()
```

```python
plt.plot(df['bedrooms'],df['bathrooms'])
plt.xlabel('Bedrooms')
plt.ylabel('Bathrooms')
plt.title("Ratio of bedrooms to bathrooms")
# plt.savefig("line_graph.png")
```

```python
index=[0,1,2,3,4,5]
c=['green','red',"yellow","blue","black","orange"]
plt.bar(index,df['stories'][0:6],color=c)
plt.xlabel("stories")
plt.ylabel("index")
plt.title("No.of Stories for first six entries")

# plt.savefig("Bar.png")
plt.show()
```

```python
#scatter plot


c=['orange','green','yellow','red']
size=[500,500,500,500]
# plt.scatter(x,y,color=c,sizes=size)
plt.scatter(df['bedrooms'][:4],df['stories'][:4],color=c,sizes=size,alpha=0.5)
plt.colorbar()
plt.title("Bedrooms and Stories")
plt.xlabel("Bedrooms")
plt.ylabel("stories")



plt.text(df['bedrooms'][0], df['stories'][0], "Index-1", fontsize=9, ha='right')
plt.text(df['bedrooms'][1], df['stories'][1], "Index-2", fontsize=9, ha='right')
plt.text(df['bedrooms'][2], df['stories'][2], "Index-3", fontsize=9, ha='left')
plt.text(df['bedrooms'][3], df['stories'][3], "Index-4", fontsize=9, ha='right')
#plt.savefig("scatterplot.png")
```

```python
# pie chart 

ex=[0,0.5,0,0,0]
plt.pie(df['area'][:5],labels=df['price'][:5],explode=ex,labeldistance=0.5,radius=1.5,autopct='%1.2f%%')
# plt.savefig('pie_chart.jpeg')
```

```python
plt.hist(df['area'], bins=30, edgecolor='black', align="mid")  
plt.title("Histogram of Area")
plt.xlabel("Area")
plt.ylabel("Frequency")
plt.savefig("histogram_area.png")
plt.show()
```

```python
plt.hist(df['price'], bins=30, edgecolor='black', align="mid")   
plt.title("Histogram of Price")#
plt.xlabel("Price")
plt.ylabel("Frequency")
# plt.savefig("histogram_price.png")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x="basement",data=df)
# sns.countplot(y="Product line",data=sales)
# plt.savefig("seaborn_fig.png")
```

```python
price_data = df['price'][:10].values.reshape(-1, 1)

sns.heatmap(price_data, annot=True, cmap='plasma', cbar=True)
plt.title("Heatmap of Price (First 10 Rows)")
plt.xlabel("Price")
#plt.savefig("heatmap.jpeg")
plt.show()
```
